```python
usernewbook_df = newbook_df.pivot_table(index='UserID',
                                  columns='BookTitle',
                                  values='BookRating')
```

```python
usernewbook_df
```

```python
# Impute thos NaNs with 0 values
usernewbook_df.fillna(0, inplace = True)
usernewbook_df
```

```python
#Caluculating Cosine similarity between user
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1-pairwise_distances(usernewbook_df.values, metric = 'cosine')
user_sim
```

```python
# Store the results in a Dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
# Set the index and column names to user IDs
user_sim_df.index = newbook_df['UserID'].unique()
user_sim_df.columns = newbook_df['UserID'].unique()
```

```python
user_sim_df.iloc[0:100, 0:100]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:100, 0:100]
```

```python
# Most Similar users
user_sim_df.idxmax(axis=1)[0:5]
```
