## Independent variable (Only Numerical Data)

```python
df.data = df.iloc[:,1:]
df.data.head()
```

```python
# Converting into numpy array
DF = df.data.values
DF
```

```python
df.data
```

```python
# Normalizing the numerical data
from sklearn.preprocessing import scale
df_normal = scale(DF)
df_normal
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(df_normal)
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative Variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color='red')
```

```python
pca_values[:,0:3]
```

```python
#Final Dataframe
final_df = pd.concat([df['Type'], pd.DataFrame(pca_values[:,0:3], columns=['pc1','pc2','pc3'])], axis=1)
final_df
```

```python
#Visualization of PCAs
import seaborn as sns
fig=plt.figure(figsize=(18,18))
sns.scatterplot(data=final_df);
```

```python
sns.scatterplot(data=final_df, x='pc1', y='pc2', hue = 'Type')
```

```python
# As we have already normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(df_normal, method = 'ward'))
```

```python
y=pd.DataFrame(y_hc, columns=['Cluster ID'])
y['Cluster ID'].value_counts()
```

```python
# Adding Clusters to Dataset
df1=df.copy()
df1['Cluster ID'] = hc.labels_
df1
```

#K-Means Clustering 

```python
from sklearn.cluster import KMeans
```

```python

# As we already have normalized data
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
#Within-Cluster sum of Squares Criterion 
wcss = []
for i in range(1,8):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 200, n_init = 7, random_state = 0)
    kmeans.fit(df_normal)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,8), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```
